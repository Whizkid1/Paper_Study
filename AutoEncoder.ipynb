# Autoencoder 모델 및 여러 방식

autoencoder훈련이 나왔을 때 Batchnormalization, initialization 기법이 없었을 때다. 그래서 pre-training을 Stacking autoencoder 방식을 사용했다.(stacking은 weight을 초기화하는 방식)
1. Denoising autoencoder: random noise를 추가해서 학습 상승
    - 학습이 진행할 수록 high level feature을 추출
    - data에 error을 추가해도 manifold에서는 원 데이터에 같은 곳에 mapping
2. Stacked denoising autoencoder: Zero-masking noise를 변경시켜가면서 학습을 해보면 noise를 추가
    - minist에서 25% noise 추가 시 loss 값이 가장 낮음
3. Stochastic contractive autoencoder: noise를 추가를 Loss function의 regularizer텀으로 넣어주는 것(여기서 부터 모르겠다...)
3. Contractive autoencoder: SCAE 의 stochastic을 deterministic 텀으로 넣어준 것(하...이것도 공부를 더...)
