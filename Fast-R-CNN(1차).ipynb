# 개선하려는 점

1. RoI (Region of Interest) 마다 CNN연산을 함으로써 속도가 저하되는 점
2. multi-stage pipelines으로써 모델을 한번에 학습시키지 못하는 점
    - bottleneck 현상을 개선
3. backpropagation이 안되는 점


- R-CNN
    - Extract image regions
    - 1 CNN per region(2000 CNNs)
    - Classify region-based features
    - Complexity: ~224 x 224 x 2000
- Fast R-CNN
    - 1 CNN on the entire image
    - Extract features from feature map regions
    - Classify region-based features
    - Complexity: ~600 x 1000 x 1
    - ~160x faster than R-CNN

# RoI Pooling layer

1. FCL를 지나기 전에 CNN을 통해 추출된 feature을 region of interest(RoI)영역을 max-pooling을 통해 저차원의 벡터로 축소
2. output들의 크기를 전부 동일한 크기로 맞춘다.
    - FCL을 지나기 위해서는 크기가 일정해야한다.
    - SPPnet에서 사용한 SPP layer의 하나의 pyramid level만을 이용한 특수 현상
    - 저차원으로 매핑하기 위해 HxW로의 max pooling(H와 W는 hyperparameter)
        1. 미리 설정한 HxW크기로 만들어주기 위해서 (h/H) * (w/H) 크기만큼 grid를 RoI위에 만든다.
        2. RoI를 grid크기로 split시킨 뒤 max pooling을 적용시켜 결국 각 grid 칸마다 하나의 값을 추출

3. RoI는 직사각형 모양을 띄며 (r,c,h,w)의 튜플 형태
    - r, c : 위, 왼쪽 코너를 의미
    - h, w : 높이와 너비를 의미

- 이를 통해 중요한 feature을 남겨두고, 계산 속도를 줄이는 방향으로 모델을 업그레이드 시켰다.
